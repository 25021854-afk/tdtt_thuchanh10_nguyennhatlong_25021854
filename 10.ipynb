w9a1 Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào.

```python
list2 = input().split()
list1 = input().split()

def merge_sorted_lists(L1, L2):
    result = []
    i, j = 0, 0
    
    while i < len(L1) and j < len(L2):
        if L1[i] <= L2[j]:
            result.append(L1[i])
            i += 1
        else:
            result.append(L2[j])
            j += 1

    while i < len(L1):
        result.append(L1[i])
        i += 1

    while j < len(L2):
        result.append(L2[j])
        j += 1
        
    return result

merged = merge_sorted_lists(list1, list2)
print(*(merged))
```

w9a2 Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể.

```python
def count_less_equal(matrix, target, n, m):
    count = 0
    row, col = n - 1, 0
    while row >= 0 and col < m:
        if matrix[row][col] <= target:
            count += (row + 1)
            col += 1
        else:
            row -= 1
    return count

def find_kth_element(matrix, k):
    n, m = 3, 5
    low = matrix[0][0]
    high = matrix[n-1][m-1]
    ans = low
    
    while low <= high:
        mid = (low + high) // 2
        # Đếm số lượng phần tử nhỏ hơn hoặc bằng mid
        if count_less_equal(matrix, mid, n, m) >= k:
            ans = mid
            high = mid - 1
        else:
            low = mid + 1
    return ans

# --- NHẬP INPUT ---
matrix = []
for i in range(3):
    row = list(map(int, input().split()))
    matrix.append(row)

k = int(input())

# --- XUẤT OUTPUT ---
print(find_kth_element(matrix, k))
```

w9a4 Cho xâu a gồm các chữ cái thường và in hoa và xâu b chỉ gồm các chữ cái in hoa. Ta có thể biến đổi xâu a theo hai bước như sau: -Chuyển một số chữ cái thường trong xâu a thành chữ cái in hoa (có thể không biến đổi chữ cái nào) -Xóa hết tất cả các chữ cái thường trong xâu a. Cho 2 xâu a,b, xác định xem có cách biến đổi từ xấu a sang xâu b không. Ghi ra "YES" nếu có và "NO" nếu ngược lại. Đầu vào: Dòng đầu tiên chứa 1 số nguyên Q là số lượng câu hỏi. Tương ứng với mỗi câu hỏi, đưa ra 2 xâu a,b trên 2 dòng khác nhau.(Q <= 10, |a|,|b| <= 1000). Đầu ra: Với mỗi câu hỏi, ghi ra kết quả tương ứng trên một dòng.

```python
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]

    dp[0][0] = True

    for i in range(n):
        for j in range(m + 1):
            if dp[i][j]:
                if j < m and a[i].upper() == b[j]:
                    dp[i+1][j+1] = True
                
                if a[i].islower():
                    dp[i+1][j] = True

    return "YES" if dp[n][m] else "NO"

q_input = input()
if q_input.strip():
    Q = int(q_input)
    
    for _ in range(Q):
        str_a = input().strip()
        str_b = input().strip()
        
        print(can_transform(str_a, str_b))
```

w9a5 Cho hai mảng số nguyên list1 và list2, viết hàm kiểm tra xem list1 có thuộc list2 hay không (có phải là dãy con liên tiếp của list2 hay không)?

```python
def is_sublist(list1, list2):
    n1 = len(list1)
    n2 = len(list2)
    
    if n1 == 0:
        return True
    if n1 > n2:
        return False
    
    for i in range(n2 - n1 + 1):
        if list2[i : i + n1] == list1:
            return True
            
    return False

line1 = input("Nhập list1 (cách nhau bởi dấu cách): ").split()
l1 = [int(x) for x in line1]

line2 = input("Nhập list2 (cách nhau bởi dấu cách): ").split()
l2 = [int(x) for x in line2]

if is_sublist(l1, l2):
    print("YES")
else:
    print("NO")
```

w9a6 [hoán vị tiếp theo] Cho một dãy số gồm n chữ số là một hoán vị của dãy 1,2,...,n. Viết chương trình in ra hoán vị ở vị trí tiếp theo (danh sách các hoán vị được sắp xếp theo thứ tự từ điển) của hoán vị đã cho. Đầu vào của bàn phím gồm 2 dòng: -Dòng đầu tiên chứa số nguyên n -Dòng tiếp theo chứa n số nguyeenlaf một hoán vị của dãy 1,2,...,n (n <= 105) Đầu ra: in ra màn hình hoán vị tiếp theo của hoán vị đã cho.

```python
def next_permutation():
    line1 = input().strip()
    if not line1:
        return
    n = int(line1)

    line2 = input().strip()
    if not line2:
        return
    a = list(map(int, line2.split()))

    i = n - 2
    while i >= 0 and a[i] >= a[i + 1]:
        i -= 1
    
    if i >= 0:
        j = n - 1
        while a[j] <= a[i]:
            j -= 1

        a[i], a[j] = a[j], a[i]

    left = i + 1
    right = n - 1
    while left < right:
        a[left], a[right] = a[right], a[left]
        left += 1
        right -= 1
    
    print(*(a))

if __name__ == "__main__":
    next_permutation()
```

w9a7 [CanPlaceFlowers] Viết hà canplaceflower(flowerbed, k) kiểm tra xem có thể trồng hết số hoa mới được cấp vào trong luống dất của bạn hay không. Hàm nhận đầu vào là một dnah sách (list) flowerbed, đánh dấu những vị trí trong luống đất đã được trồng cây xanh trước đó (giá trị 1 ứng với vị trí cây đã được trồng và 0 ứng với vị trí đất trống) và số lượng hoa được phát thêm k (int). Hàm trả về True nếu có thể trồng hết toàn bộ k cây hoa vào luống đất và False trong trường hợp ngược lại.

```python
def can_place_flowers(flowerbed, k):
    count = 0
    n = len(flowerbed)
    
    for i in range(n):
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i - 1] == 0)
            right_empty = (i == n - 1) or (flowerbed[i + 1] == 0)
            
            if left_empty and right_empty:
                flowerbed[i] = 1
                count += 1

                if count >= k:
                    return True
                    
    return count >= k

line = input("Nhập luống đất (0 và 1, cách nhau bởi dấu cách): ").split()
flowerbed_input = [int(x) for x in line]

k_input = int(input("Nhập số lượng hoa k: "))

print(can_place_flowers(flowerbed_input, k_input))
```

w9a8 [phát hiện đường đi lặp lại của rô bốt] Ban đầu, một con rô bốt ở vị trí (0,0) trên một lưới 2D. Cho một chuỗi 'moves' chứa các thao tác di chuyển của rô bốt. Các bước di chuyển được biểu diễn bởi các ký tự: 'R' 'L' 'U' 'D'. Viết hàm def is_path_crossing(moves) Hàm nhận đầu vào là một chuỗi moves. Hàm trả về True nếu trong quá trình di chuyển rô bốt đi vào một ô mà nó đã từng đi qua trước đó. Hàm trả về False nếu rô bốt không bao giờ đi lặp lại (luôn đi vào ô mới). Lưu ý Vị trí ban dầu (0,0) được tính là tị trí đầu tiên đã đi qua.

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    
    for move in moves:
        if move == 'U':
            y += 1
        elif move == 'D':
            y -= 1
        elif move == 'L':
            x -= 1
        elif move == 'R':
            x += 1
        
        current_pos = (x, y)
        
        if current_pos in visited:
            return True
        
        visited.add(current_pos)
        
    return False

moves_input = input("Nhập chuỗi di chuyển (ví dụ: NESW hoặc UDRL): ").upper()
print(is_path_crossing(moves_input))
```

w9a9 [Tìm trung vị của 5 số] Viết hàm def Median(nums). Hàm nhận đầu vào là một dnah sách (list) nums được đảm bảo có đúng 5 số nguyên. Yêu cầu: Tìm và trả về số trung vị (số lớn thứ ba) của 5 số này. Ràng buộc: 1.Chỉ được sử dụng tối đa 6 phép so sánh, 2.Không được dùng bất kỳ hàm sắp xếp nào. Chú ý: Bạn chỉ cần viết hàm Median theo yêu cầu. Phần đọc input và gọi hàm (hàm main) đã có sẵn.

```python
def Median(nums):
    a, b, c, d, e = nums

    if a > b: a, b = b, a
    if c > d: c, d = d, c
    
    if a < c:
        a = e
        if a > b: a, b = b, a
    else:
        c = e
        if c > d: c, d = d, c

    if a < c:
        if b < c:
            return c
        else:
            return b if b < d else d
    else:
        if a > d:
            return d
        else:
            return a if a < b else b

if __name__ == "__main__":
    line = input().split()
    if len(line) == 5:
        input_nums = [int(x) for x in line]
        print(Median(input_nums))
```

w9a10 [Tìm bán kính lò sưởi nhỏ nhất] Tất cả các lò sưởi đều có cùng một bán kính r. Một ngôi nhà tại vị trí h được sưởi ấm bởi một lò sưởi tại vị trí t nếu |h-t| <= r. Viết hàm def findRadius(house, heaters). Hàm nhận đầu vào là hai danh sách (list) các số nguyên: 'house': chứa vị trí các ngôi nhà ; 'heaters': chứa vị trí của các lò sưởi. Hàm trả về một số nguyên là bán kính r nhỏ nhất cần thiết. Lưu ý: Số lượng nhà và lò sưởi là một số không âm và không vượt quá 25000, Vị trí của các ngôi nhà và lò sưởi là một số không âm và không vượt quá 10^9.

```python
import bisect

def findRadius(houses, heaters):
    heaters.sort()
    max_radius = 0
    
    for house in houses:
        idx = bisect.bisect_left(heaters, house)

        dist_right = heaters[idx] - house if idx < len(heaters) else float('inf')
        
        dist_left = house - heaters[idx - 1] if idx > 0 else float('inf')
        
        current_min_dist = min(dist_left, dist_right)
        
        max_radius = max(max_radius, current_min_dist)
        
    return max_radius

if __name__ == "__main__":
    line1 = input().split()
    houses = [int(x) for x in line1]
    
    line2 = input().split()
    heaters = [int(x) for x in line2]
    
    print(findRadius(houses, heaters))
```
